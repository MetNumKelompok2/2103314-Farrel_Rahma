```python
# Nama  : Farrel Rahma Aliyya
# NIM   : 2103314
# Akar Newton Raphson

# Definisi Fungsi
def f(x):
    return x**3 - 6*x +2 

# Definisi ungsi turunan
def g(x):
    return 3*x**2 -6

# Implementasi Metode Newton Raphson

def newtonRaphson(x0,e,N):
    print('\n\nIMPLEMENTASI METODE NEWTON RAPHSON')
    step = 1
    flag = 1
    condition = True
    while condition:
        if g(x0) == 0.0:
            print('Divide by zero error!')
            break
        
        x1 = x0 - f(x0)/g(x0)
        print('Iterasi-%d, x1 = %0.6f dan f(x1) = %0.6f' % (step, x1, f(x1)))
        x0 = x1
        step = step + 1
        
        if step > N:
            flag = 0
            break
        
        condition = abs(f(x1)) > e
    
    if flag==1:
        print('\nRequired root is: %0.8f' % x1)
    else:
        print('\nNot Convergent.')


# Input Section
x0 = input('masukkan tebakan: ')
e = input('toleransi galat: ')
N = input('batas iterasi: ')

# mengubah x0 and e to float
x0 = float(x0)
e = float(e)

# mengubah N to integer
N = int(N)

# memanggil fungsi Newton Raphson 
newtonRaphson(x0,e,N)
```

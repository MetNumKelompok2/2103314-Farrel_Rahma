**Interpolasi Lagrange**

```python
# Nama  : Farrel Rahma Aliyya
# NIM   : 2103314
# Interpolasi Lagrange

import numpy as np
    
# tampilkan grafik 
import matplotlib.pyplot as plt

# input jumlah titik
n = int(input('input jumlah titik: '))

# inisialisasi x
x = np.zeros((n))
# inisialisasi y
y = np.zeros((n))

# input data x dan y
print('input data x dan y: ')
#perulangan untuk input data
for i in range(n):
    x[i] = float(input( 'x['+str(i)+']=')) # input x[i]
    y[i] = float(input( 'y['+str(i)+']=')) # input y[i]

# input masukan yang dicari
xp = float(input('input yang ingin dicari: '))

# set data dari 0
yp = 0 

# algoritma Lagrange
for i in range(n): 
    p = 1
    for j in range(n):
        if i != j:
            p = p * (xp - x[j])/(x[i] - x[j])
    yp = yp + p * y[i]    

# tampilkan hasil
print('Pada x %.3f mempunyai f(x) atau y  %.3f' % (xp, yp))
plt.plot(x, y, marker = 'x')
plt.plot(xp, yp, marker = 'o')
plt.show()
```
